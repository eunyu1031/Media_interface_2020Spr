## 2. 데이터 클리닝: 중복과 누락
- 기본정보에서 확인했을 때 누락된 데이터는 없음
- 중복된 데이터는 groupby로 해결

## 3. 데이터 클리닝: 그 외 오타/오류/서식/자료형 문제 
- 매출액점유율(%)을 float로 변경

```python
data['매출액점유율(%)'] = data['매출액점유율(%)'].str.replace('%', '').astype("float")
```

```python
data.info()
```

중복된 영화 코드 확인

```python
data[data.duplicated('영화코드', keep = False)].sort_values('영화코드')
```

영화 코드를 기준으로 groupby-sum한 결과를, 영화 코드를 기준으로 merge

```python
temp = pd.merge(data.iloc[:, [0, 2, 3]], data.groupby('영화코드').sum().iloc[:, 1:], on = '영화코드')
temp[temp.duplicated('영화코드', keep = False)].sort_values('영화코드')
```

중복된 영화 코드가 같은 값을 가지고 있으므로 둘 중 하나를 버리고 결과를 확인

```python
temp.drop_duplicates('영화코드', inplace = True)
```

```python
temp[temp.duplicated('영화코드',keep = False)]
```

```python
vis = temp.reset_index(drop = True)
```

#### (1) 529개 영화의 매출액은 어떤 분포를 그리는가? 
- A: 기술통계에서 예상할 수 있다시피 일부의 영화만 매우 높은 매출을 올리고 대부분은 그저 그렇다.

#### (2) 영화의 월별 개봉 분포는 어떠한가?
- A: 방학과 연말연시, 휴가 기간을 포함하는 12월과 1월, 7-8월에 개봉이 많은 편이다.

```python
vis['개봉연도'] = vis['개봉연월일'].str.split('-', expand = True).iloc[:, 0].astype('int64')
vis['개봉월'] = vis['개봉연월일'].str.split('-', expand = True).iloc[:, 1].astype('int64')
vis.head()
```

```python
fig, ax = plt.subplots()

sns.countplot(x = '개봉월', 
            data = vis[vis['개봉연도']<2020],
             ax = ax)

ax.set(title = '월별 개봉한 영화의 수', ylabel = '개봉영화 수')
plt.show()
```

#### (3) 각 월별 영화 수요는 어떠한가?
- A: 7-8월에 많은 편이나 큰 성공을 거둔 영화는 3월을 제외하고 크게 시즌을 타지 않는다.

```python
fig, ax = plt.subplots()

sns.boxplot(x = '개봉월',
            y = "관객수",
            data = vis[vis['개봉연도']<2020], 
            ax = ax)

ax.set_title("월별 관객수 분포")
plt.show()
```

#### (4) 2010년부터 한국 영화시장의 1~4월의 월별 총 매출은 어떠하였는가?
- A: 연도/월별로 부침이 있지만 2020년 3, 4월이 매출은 매우 낮다.

```python
fig, ax = plt.subplots()

sns.barplot(x = '개봉연도',
            y = '매출액',
            hue = '개봉월',
            data = vis[vis['개봉월']<5],
            estimator = np.sum,
            ci = None,
           palette = 'Paired',
           ax = ax)

ax.set_title('한국 영화시장의 1~4월의 월별 총 매출')
plt.show()
```

#### (5) 사회적 거리두기를 본격화한 2020년 4월 총 매출은 어떠하였는가? 
- A. 폭락했다.

```python
fig, ax = plt.subplots()

sns.barplot(x = '매출액',
            y = '개봉연도',
            data = vis[vis['개봉월']==4],
            estimator = np.sum,
            ci = None,
           ax = ax, 
            orient = 'h',
           color = 'gray')

ax.set_title("2010년부터 2020년까지 4월의 top 50 영화 매출 합산")
plt.show()
```

#### (6) 천만영화가 차지하는 관객수 비중은 어느 정도인가?
- A. Top 50의 개봉연도 기준으로 적게는 10% 전후에서 많게는 30%까지 차지한다.

관객수 변수를 바탕으로 천만을 나타내는 이진형 변수 생성

```python
vis['천만'] = pd.cut(vis['관객수'], [0, 9999999, 20000000], labels = [0, 1])
vis[vis['천만']== 1].head()
```
